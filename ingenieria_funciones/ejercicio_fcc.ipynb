```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Creamos los datos.
data = {
    'Age' : [25, 30, 35, 40, 45],
    'Income' : [50000, 60000, 75000, 90000, 80000],
    'Education' : ['High School', 'Bachelor', 'Master', 'Ph.D.', 'Bachelor'],
    'City' : ['New York', 'San Francisco', 'Chicago', 'Los Angeles', 'Miami'],
    'Gender' : ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Productivity' : [5, 4, 3, 2, 4]
}
df = pd.DataFrame(data)
df
```

```python
# One-hot encoding.
df_encoded = pd.get_dummies(df, columns=['Education', 'City', 'Gender'])
df_encoded
```

```python
# Inicializamos el Standar Scaler.
scaler = StandardScaler()
# Ajustamos y transformamos las columnas 'Age' y 'Income'.
df_encoded[['Age', 'Income']] = scaler.fit_transform(df_encoded[['Age', 'Income']])
df_encoded
```

```python
# Creamos la nueva función 'Ingresos por edad'.
df_encoded['Income per Age'] = df_encoded['Income'] / df_encoded['Age']
df_encoded
```

```python
# Separamos la variable objetivo y las características.
X = df_encoded.drop('Productivity', axis=1)
y = df_encoded['Productivity']

# Inicializamos el modelo de regresión linear.
model = LinearRegression()

# Realizamos RFE con validación cruzada.
rfe = RFE(model, n_features_to_select=3) #
fit = rfe.fit(X, y) # 

# Imprimimos las características seleccionadas.
selected_features = X.columns[fit.support_]
print('Selected Features:', selected_features)
```
